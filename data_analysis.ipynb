```python
import pandas as pd
import os

# Path to the Excel file
excel_path = "/projects/b1198/epifluidlab/ravi/rhinology_reimbursment/raw_data/rhinology_reimbursement_cpt_codes.xlsx"

# Directory where the Excel file is located
output_dir = os.path.dirname(excel_path)

# Load the Excel file
xls = pd.ExcelFile(excel_path)

# Loop through each sheet and save as CSV
for sheet_name in xls.sheet_names:
    df = pd.read_excel(xls, sheet_name=sheet_name)
    
    # Create CSV path using sheet name
    csv_path = os.path.join(output_dir, f"{sheet_name}.csv")
    
    # Save as CSV
    df.to_csv(csv_path, index=False)

    print(f"Saved: {csv_path}")
```
